### Step 4:  Feature engineering

```python
not_fraud = df[df.Class == 0]
fraud = df[df.Class == 1]
```

### Assumption 2 : No multicolinearity

```python
df1 = new_df.drop("Class",axis = 1)
df1
```

```python
vif_list = []
for i in range(df1.shape[1]):
    vif = variance_inflation_factor(df1.to_numpy(),i)
    vif_list.append(vif)
vif_list
```

```python
vif_df = pd.DataFrame()
vif_df["Features"] = df1.columns
vif_df
```

```python
vif_df["VIF"] = vif_list
vif_df
```

```python
x = df1
y = new_df["Class"]
y.head(8)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state = 10)
```

### Step 6: Model Training

```python
log = LogisticRegression()
log_model = log.fit(xtrain,ytrain)
log_model
```

```python
ytrain_pred = log_model.predict(xtrain)
(ytrain_pred,ytrain)
```
